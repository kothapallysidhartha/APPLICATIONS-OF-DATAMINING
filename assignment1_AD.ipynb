# 1.	Load the dataset using Pandas. **bold text**


```python
import pandas as pd
file_path = "student-dataset.csv"
df = pd.read_csv(file_path)
```

```python
df
```

# **	Fill missing science scores using median imputation.**

```python
df.isnull()
```

```python

print("Missing values in each column:")
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sciences_median = df['sciences.grade'].median()
df['sciences.grade'].fillna(sciences_median, inplace=True)
```

# **Columns for which we need to perform Normalization**

```python
exam_columns = ['english.grade', 'math.grade', 'sciences.grade', 'language.grade']
```

```python
scaler = MinMaxScaler()
df[exam_columns] = scaler.fit_transform(df[exam_columns])
```

```python
df[exam_columns]
```

```python
df.to_csv("updated_student_dataset.csv", index=False)
```
